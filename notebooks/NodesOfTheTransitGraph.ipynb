<h1> Examining the GPS Locations </h1>

```python
import pandas as pd
import numpy as np
import json
import pylab as pl
```

```python
pl.rcParams.update({'font.size':16})
pl.style.use('seaborn-colorblind')
```

```python
# Get MTA Station locations.
mta_stations = pd.read_csv('../data/mta/mta_stations.csv')

# Get Citibike Station locations.
citibike_stations = pd.read_csv('../data/citibike/citibike_stations.csv')
```

```python
# Clean mta station names.
mta_stations = mta_stations.rename(columns={'GTFS Latitude':'latitude', 'GTFS Longitude':'longitude', 'Daytime Routes':'Routes'})
```

```python
mta_stations.Division.unique()
```

```python
print('There are {} citibike stations and {} MTA stations.'.format(citibike_stations.shape[0], mta_stations.shape[0]))
```

```python
fig, ax = pl.subplots(figsize=(12,12))
ax.scatter(citibike_stations.longitude, citibike_stations.latitude)
ax.scatter(mta_stations.longitude, mta_stations.latitude, color='orange')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude');
```
